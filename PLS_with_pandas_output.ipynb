```python
dataFrame = pd.read_excel('decimated_spectra.xlsx')
display(dataFrame)
```

### Define the primary ID (index) of the table as the column named sample

```python
named_Obs_dataFrame = dataFrame.set_index('sample')
named_Obs_dataFrame
```

### Get the selection of the data table designated as X-data

```python
data = named_Obs_dataFrame.loc['fm01':'fm10' , '250':'418']
data
```

```python
test_set = named_Obs_dataFrame.loc['fm11':'fm12' , '250':'418']
test_set 
```

### Plot the X-data using the pandas plotting tools

```python
raw_plt_axes = data.transpose().plot()
raw_plt_axes.set_xlabel('nm')
_ = raw_plt_axes.set_ylabel('Absorbance')
```

### Get the selection of the data table designated as Y-data

```python
y = named_Obs_dataFrame.loc['fm01':'fm10' , ['Met']]
y
```

### Evaluate the possible number of components to use in the PLS model 

```python
R2, Q2 = mv.evalPLS_Q2(data, y)
print('Q2', Q2)
```

### As the Q2 values levels out at 2 components, 2 components are selected

```python
M1 = mv.PLS_model(n_components=2)
```

```python
M1.fit(data, y)
```

### Compare the measured values for the test with with the predicted

```python
test_set_y = named_Obs_dataFrame.loc['fm11':'fm12' , ['Met']]
test_set_y
```

```python
M1.predict(test_set)
```

### Check the DModX residuals per spectrum for the predicted test set 

```python
M1.DModXpred(test_set)
```

### And compare with the training set residuals

```python
M1.DModXpred(data)
```

```python
all_DModX = pd.concat((M1.DModXpred(data), M1.DModXpred(test_set)))
all_DModX
```

```python
all_DModX.plot.bar(rot=0)
```

### Summary
Although the DModX values for the testset are higher (and known to be outside the model design space), we can still see that the predicted y-values are not deviating much from the expected results. For a fully worked through case, more than two spectra are needed in the test set and those should be both inside and outside the design space of the model to enable a better foundation for a decision where to set the threshold for unacceptable spectra.
